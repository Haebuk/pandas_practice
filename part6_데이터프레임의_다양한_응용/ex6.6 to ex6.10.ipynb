# 데이터프레임에 pipe() 적용 

```python
import seaborn as sns
```

### titanic 데이터셋에서 age, fare 2개 열을 선택하여 데이터프레임 

```python
titanic = sns.load_dataset('titanic')
df = titanic.loc[:,['age','fare']]
```

### 각 열의 NaN 찾기 - 데이터프레임 전달하면 데이터프레임 반환 

```python
def missing_value(x):
    return x.isnull()
```

### 각 열의 NaN 개수 반환 - 데이터프레임 전달하면 시리즈 반환 

```python
def missing_count(x):
    return missing_value(x).sum()
```

### 데이터프레임의 총 NaN 개수 - 데이터프레임 전달하면 값 반환 

```python
def total_number_missing(x):
    return missing_count(x).sum()
```

### 데이터프레임에 pipe() 메소드로 함수 매핑 

```python
result_df = df.pipe(missing_value)
result_df.head()
```

```python
result_series = df.pipe(missing_count)
result_series
```

```python
result_value = df.pipe(total_number_missing)
result_value
```

# 예제 6-9 불린 인덱싱 

### 나이가 10대(10-19세)인 승객만 따로 선택

```python
mask1 = (titanic.age >= 10) & (titanic.age < 20)
```

```python
df_teenage = titanic.loc[mask1, :]
df_teenage.head()
```

### 나이가 10세 미만(0-9세)이고 여성인 승객 

```python
mask2 = (titanic.age < 10) & (titanic.sex == 'female')
```

```python
df_female_under10 = titanic.loc[mask2, :]
df_female_under10
```

### 나이가 10세 미만 또는 60세 이상인 승객의 age, sex, alone 

```python
mask3 = (titanic.age < 10) | (titanic.age >= 60)
```

```python
df_under10_morethan60 = titanic.loc[mask3, ['age', 'sex', 'alone']]
df_under10_morethan60
```

# 예제 6-10 isin() 필터링 

### 함께 탑승한 형제 또는 배우자의 수가 3, 4, 5인<br> 승객만 따로 추출 - 불린 인덱싱

```python
mask3 = titanic['sibsp'] == 3
mask4 = titanic['sibsp'] == 4
mask5 = titanic['sibsp'] == 5
```

```python
df_boolean = titanic[mask3 | mask4 | mask5]
df_boolean.head()
```

### isin() 메소드 활용하여 동일 조건으로 추출 

```python
isin_filter = titanic['sibsp'].isin([3, 4, 5])
df_isin = titanic[isin_filter]
df_isin.head()
```
