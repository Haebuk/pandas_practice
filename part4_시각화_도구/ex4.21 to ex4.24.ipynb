# 예제 4-21 버블 차트 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### read_csv() 함수로 df 생성 

```python
df = pd.read_csv('./auto-mpg.csv', header= None)
df.head()
```

### 열 이름 지정 

```python
df.columns = ['mpg', 'cylinders', 'displayment', 'horsepower','weight',
              'acceleration', 'model year', 'origin', 'name']
df.head()
```

### cylinder 개수의 상대적 비율을 계산하여 시리즈 생성 

```python
cylinder_size = df.cylinders/df.cylinders.max() * 300
cylinder_size.unique()
```

300을 곱한 이유: 안 곱하면 밑에 있는 버블 차트에서 매우 작게 보임 

(마치 그냥 scatter plot 같이 보이는 느낌)

### 3개의 변수로 산점도 그리기 

```python
df.plot(kind= 'scatter', x= 'weight', y= 'mpg', c= 'coral', figsize= (10, 5),
       s= cylinder_size, alpha= 0.3)
plt.title('Scatter Plot: mpg-weight-cylinders')
```

s 옵션에 cylinder_size를 입력하여 값의 크기에 따라 

점의 크기를 값에 따라 다르게 표시한다.

# 예제 4-22 그림 파일로 저장 

### 3개의 변수로 산점도 그리기 

```python
df.plot(kind= 'scatter', x= 'weight', y= 'mpg', marker= '+', figsize= (10, 5),
       cmap= 'viridis', c= cylinder_size, s= 50, alpha= 0.3)
plt.title('Scatter Plot: mpg-weight-cylinders')
plt.savefig('./scatter.png')
plt.savefig('./scatter_transparent.png', transparent= True)
```

transparent= True 옵션은 그림 배경을 투명하게 지정하는 옵션

c 옵션에 cylinder_size를 할당하여 값에 따라 다른 색상으로 표현

색상을 정하는 컬러맵(cmap)으로 'viridis' 옵션 사용

# 예제 4-23 파이 차트 

```python
plt.style.use('default')
```

### 데이터 개수 카운트를 위해 값 1을 가진 열 추가 

```python
df['count'] = 1
df_origin = df.groupby('origin').sum() # origin 열을 기준으로 그룹화, 합계 연산
df_origin.head()
```

### 제조국가(origin) 값을 실제 지역명으로 변경 

```python
df_origin.index = ['USA', 'EU', 'JPN']
```

### 제조국가(origin) 열에 대한 파이 차트 그리기-count 열 데이터 사용 

```python
df_origin['count'].plot(kind='pie', figsize=(7, 5),
                       autopct='%1.1f%%', # 퍼센트 % 표시
                       startangle=10, # 파이 조각을 나누는 시작점(각도 표시)
                       colors=['chocolate', 'bisque', 'cadetblue']) # 색상 리스트
plt.title('Model Origin', size= 20)
plt.axis('equal') # 파이 차트의 비율을 같게(원에 가깝게) 조정
plt.legend(labels=df_origin.index, loc='upper right')
```

# 예제 4-24 박스 플롯 

```python
plt.style.use('seaborn-poster') # 스타일 서식 지정
plt.rcParams['axes.unicode_minus'] = False # 마이너스 부호 출력 설정
```

### 그래프 객체 생성(figure에 2개의 서브 플롯 생성) 

```python
fig = plt.figure(figsize=(15, 5))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)
# ax 객체에 boxplot 메소드로 그래프 출력
ax1.boxplot(x=[df[df['origin']==1]['mpg'],
               df[df['origin']==2]['mpg'],
               df[df['origin']==3]['mpg']],
           labels=['USA', 'EU', 'JAPAN'])
ax2.boxplot(x=[df[df['origin']==1]['mpg'],
               df[df['origin']==2]['mpg'],
               df[df['origin']==3]['mpg']],
           labels=['USA', 'EU', 'JAPAN'],
           vert=False)
ax1.set_title('제조국가별 연비 분포(수직 박스 플롯)')
ax2.set_title('제조국가별 연비 분포(수평 박스 플롯)')
```
